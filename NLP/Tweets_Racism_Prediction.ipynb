```python
import pandas as pd
```

```python
datadf = pd.read_csv('./data/tweet/train.csv', index_col='id')
```

```python
def clean_tweet(df):
    """ Cleans tweet
    df: input data frame
    
    returns cleaned data frame
    """
    
    dfoutput = df
    dfoutput.replace(to_replace="@user", value="", regex=True, inplace=True)
    dfoutput.replace(to_replace="&amp;", value="and", regex=True, inplace=True)
    dfoutput.replace(to_replace="can\'t", value="cannot", regex=True, inplace=True)
    dfoutput.replace(to_replace="don\'t", value="do not", regex=True, inplace=True)
    dfoutput.replace(to_replace="doesn\'t", value="does not", regex=True, inplace=True)
    dfoutput.replace(to_replace="aren\'t", value="are not", regex=True, inplace=True)
    dfoutput.replace(to_replace="isn\'t", value="is not", regex=True, inplace=True)
    dfoutput.replace(to_replace="i\'m", value="i am", regex=True, inplace=True)
    dfoutput.replace(to_replace="you\'ve", value="you have", regex=True, inplace=True)
    dfoutput.replace(to_replace="she\'s", value="she is", regex=True, inplace=True)
    dfoutput.replace(to_replace="wouldn\'t", value="would not", regex=True, inplace=True)
    dfoutput.replace(to_replace="ain\'t", value="does not", regex=True, inplace=True)
    dfoutput.replace(to_replace="hasn\'t", value="has not", regex=True, inplace=True)
    dfoutput.replace(to_replace="haven\'t", value="have not", regex=True, inplace=True)
    dfoutput.replace(to_replace="weren\'t", value="were not", regex=True, inplace=True)
    dfoutput.replace(to_replace="that\'s", value="that is", regex=True, inplace=True)
    dfoutput.replace(to_replace="you\'re", value="you are", regex=True, inplace=True)
     
    dfoutput.replace(to_replace="he\'s", value="he is", regex=True, inplace=True)
    dfoutput.replace(to_replace="we\'re", value="we are", regex=True, inplace=True)
    dfoutput.replace(to_replace="it\'s", value="it is", regex=True, inplace=True)
    dfoutput.replace(to_replace="they\'re", value="they are", regex=True, inplace=True)
    dfoutput.replace(to_replace="there\'s", value="there is", regex=True, inplace=True)
    dfoutput.replace(to_replace="wasn\'t", value="was not", regex=True, inplace=True)
    dfoutput.replace(to_replace="I\'ve", value="I have", regex=True, inplace=True)
    dfoutput.replace(to_replace="we\'ve", value="we have", regex=True, inplace=True)
    dfoutput.replace(to_replace="they\'ve", value="they have", regex=True, inplace=True)
    dfoutput.replace(to_replace="there\'s", value="there has", regex=True, inplace=True)
    dfoutput.replace(to_replace="hasn\'t", value="has not", regex=True, inplace=True)
    dfoutput.replace(to_replace="haven\'t", value="have not", regex=True, inplace=True)
    dfoutput.replace(to_replace="I\'d", value="I had", regex=True, inplace=True)
    dfoutput.replace(to_replace="you\'d", value="you had", regex=True, inplace=True)
    dfoutput.replace(to_replace="he\'d", value="he had", regex=True, inplace=True)
    dfoutput.replace(to_replace="she\'d", value="she had", regex=True, inplace=True)
    dfoutput.replace(to_replace="it\'d", value="it had", regex=True, inplace=True)
    dfoutput.replace(to_replace="they\'d", value="they had", regex=True, inplace=True)
    dfoutput.replace(to_replace="we\'d", value="we had", regex=True, inplace=True)
    dfoutput.replace(to_replace="I\'ll", value="I will", regex=True, inplace=True)
    dfoutput.replace(to_replace="you\'ll", value="you will", regex=True, inplace=True)
    dfoutput.replace(to_replace="he\'ll", value="he will", regex=True, inplace=True)
    dfoutput.replace(to_replace="she\'ll", value="she will", regex=True, inplace=True)
    dfoutput.replace(to_replace="we\'ll", value="we will", regex=True, inplace=True)
    dfoutput.replace(to_replace="they\'ll", value="they will", regex=True, inplace=True)
    dfoutput.replace(to_replace="there\'ll", value="there will", regex=True, inplace=True)

    dfoutput.replace(to_replace="couldn't", value="could not", regex=True, inplace=True)
    dfoutput.replace(to_replace="won\'t", value="will not", regex=True, inplace=True)
    dfoutput.replace(to_replace="shouldn\'t", value="should not", regex=True, inplace=True)
    dfoutput.replace(to_replace="hadn\'t", value="had not", regex=True, inplace=True)

    
    dfoutput.replace('[^a-zA-Z ]', ' ' , regex=True, inplace=True)
    
    return dfoutput
```

```python
clean_data = clean_tweet(datadf)
```

```python
from keras.preprocessing.text import Tokenizer
```

```python
t = Tokenizer()
```

```python
X = clean_data.drop('label', axis=1)
y = clean_data['label']
```

```python
# split data into test and train
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
onehot_y_train = pd.get_dummies(y_train).as_matrix()
onehot_y_test = pd.get_dummies(y_test).as_matrix()
```

```python
# import for Keras
```

```python
from keras.layers import Dense, Activation, Embedding, LSTM, Dropout
```

```python
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
```

```python
# create embedding layer
```

```python
def convert_word_to_indices_pad(X, max_length):
    """ converts a word into a vector of integers
    
    X -> an array of sentences
    max_length -> max_length of sentence for padding purposes
    
    returns a tuple of tokenzier and padded vector of integers
    """
   
    X_indices = t.texts_to_sequences(X)
    
    # return tokenizer and padded vector of indices
    return pad_sequences(X_indices, maxlen=max_length, padding='post')
```

```python
def get_max_len(X):
    """ array of sentences in X, return max length of sentence"""
    
    max_len = 0
    for i in range(len(X)):
        
        len1 = len(X.iloc[i].split())
        
        if len1 > max_len:
            max_len = len1
            
    return max_len
```

```python
max_len = get_max_len(X_train['tweet'])
```

```python
X_train_indices = convert_word_to_indices_pad(X_train['tweet'], max_len)
```

```python
# create embedding matrix to be used in Keras embedding layer
```

```python
def create_embedding_matrix(t, model):
    """ Create embedding matrix to be passed to Keras Embedding layer.
    t -> tokenizer
    model -> glove model
    
    returns a matrix of shape vocab_size, glove_vector_length
    """
    
    vocab_size = len(t.word_index) + 1
    
    
    e_matrix = np.zeros((vocab_size, glove_vector_length))
    
    for word in t.word_index:
        
        if model.vocab.get(word) is not None:
            e_matrix[t.word_index[word]] = model[word]
    
    return e_matrix
```

```python
e_matrix = create_embedding_matrix(t, tmodel)
```

```python
vocab_size = len(t.word_index) + 1
```

```python
num_outputs = 2 #classes
```

```python
model = Sequential()
```

```python
model.add(Embedding(vocab_size, glove_vector_length, trainable=False, weights=[e_matrix]))
```

```python
model.add(LSTM(128, return_sequences=True))
model.add(Dropout(0.5))
model.add(LSTM(128, return_sequences=False))
model.add(Dropout(0.5))
model.add(Dense(num_outputs))
model.add(Activation('softmax'))
```

```python
model.compile(optimizer='adam', 
             loss='categorical_crossentropy',
             metrics=['accuracy'])
```

```python
# evaluate and predict
```

```python
model.fit(X_train_indices, onehot_y_train,epochs=4, batch_size=32, shuffle=True )
```

```python
X_test_indices = convert_word_to_indices_pad(X_test['tweet'], get_max_len(X_test['tweet']))
```

```python
loss, acc = model.evaluate(X_test_indices, onehot_y_test)

print("test accuracy: ", acc)
```

```python
pred = model.predict(X_test_indices)
```

```python
from sklearn.metrics import classification_report
```

```python
predictions = np.argmax(pred, axis=1)
```

```python
print(classification_report(predictions, y_test))
```
