```python
days = [1,2,3,4,5]

sleeping = [9,6,7,8,10]
eating = [2,3,4,2,2]
working = [10,12,8,10,9]
playing = [3, 3, 5, 4, 3]

plt.plot([],[], label="sleeping", color='m', linewidth=5)
plt.plot([],[], label="eating", color='c', linewidth=5)
plt.plot([],[], label="working", color='r', linewidth=5)
plt.plot([],[], label="playing", color='g', linewidth=5)

plt.stackplot(days, sleeping,eating,working,playing, colors=["m","c","r","g"])

plt.xlabel("days")
plt.ylabel("activity")
plt.title("Activity by days")

plt.legend()
```

```python
import pandas as pd 
from matplotlib.finance import candlestick_ohlc
import matplotlib.dates as mdates
import matplotlib.ticker as mticker
from matplotlib import style
```

```python
dateparse = lambda x: pd.datetime.strptime(x, '%d-%m-%Y')
quotes = pd.read_csv("data/stock_price.csv", parse_dates=['Date'], index_col="Date", date_parser=dateparse)
```

```python
quotes
```

```python
style.use('fivethirtyeight')

fig = plt.figure()

ax = fig.add_subplot(111)

candlestick_ohlc(ax, zip(mdates.date2num(quotes.index.to_pydatetime()),
                         quotes['Open'], quotes['High'],
                         quotes['Low'], quotes['Close']),
                 width=0.4,
                colordown="#db3f3f",
                colorup="#77d879")

ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m-%d"))
ax.xaxis.set_major_locator(mticker.MaxNLocator(10))

plt.xlabel("time")
plt.ylabel("stock price")

ax.xaxis.label.set_color('c')
ax.yaxis.label.set_color('g')

plt.xticks(rotation=45)


plt.title("Candlestick graph for stock price")
```

```python
def create_data():
    xs = []
    ys = []
    
    for i in range(10):
        xs.append(i)
        ys.append(random.randrange(10))
        
    return xs, ys
        
        
```

```python
def moving_averages(values, window):
    
    weights = np.repeat(1.0, window)/window
    smas = np.convolve(values, weights, 'valid')
    
    return smas

def high_minus_low(highs, lows):
    return highs-lows
```

```python
MA1 = 2
MA2 = 7

#style.use('fivethirtyeight')

fig = plt.figure()

ax1 = plt.subplot2grid((6,1), (0,0), rowspan=1, colspan=1)
plt.title("Candlestick graph for stock price")
plt.ylabel('H-L')

ax2 = plt.subplot2grid((6,1), (1,0), rowspan=4, colspan=1)
plt.ylabel("stock price")

ax3 = plt.subplot2grid((6,1), (5,0), rowspan=1, colspan=1)
plt.xlabel("time")
plt.ylabel("MAvgs")

candlestick_ohlc(ax2, zip(mdates.date2num(quotes.index.to_pydatetime()),
                         quotes['Open'], quotes['High'],
                         quotes['Low'], quotes['Close']),
                 width=0.4,
                colordown="#db3f3f",
                colorup="#77d879")



bbox_props = dict(boxstyle='larrow', fc='w', ec='k', lw=1)
ax2.annotate(str(quotes['Close'][-1]), (quotes.index[-1], quotes['Close'][-1]), 
            xytext = (quotes.index[-1]+pd.DateOffset(days=4), quotes['Close'][-1]), 
           bbox=bbox_props)

#ax2.xaxis.label.set_color('c')
ax2.yaxis.label.set_color('g')


start = len(quotes.index[MA2-1:])
ma1 = moving_averages(quotes['Close'], MA1)
ma2 = moving_averages(quotes['Close'], MA2)

ax3.plot(quotes.index[-start:], ma1[-start:], linewidth=1)
ax3.plot(quotes.index[-start:], ma2[-start:], linewidth=1)


ax3.fill_between(quotes.index[-start:], ma2[-start:], ma1[-start:],
               where = ( ma2[-start:] > ma1[-start:]),
                facecolor='r', edgecolor='r', alpha=0.5)

ax3.fill_between(quotes.index[-start:], ma2[-start:], ma1[-start:],
                where = ( ma2[-start:] < ma1[-start:]),
                facecolor='g', edgecolor='g', alpha=0.5)

ax3.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m-%d"))
ax3.xaxis.set_major_locator(mticker.MaxNLocator(10))

h_l = list(map(high_minus_low, quotes['High'], quotes['Low']))

ax1.plot_date(quotes.index, h_l, '-')
ax1.yaxis.set_major_locator(mticker.MaxNLocator(nbins=2, prune='lower'))

plt.setp(ax1.get_xticklabels(), visible=False)
plt.setp(ax2.get_xticklabels(), visible=False)

plt.xticks(rotation=45)


```
